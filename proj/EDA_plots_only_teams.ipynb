```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the teams.csv file
teams = pd.read_csv("basketballPlayoffs/teams.csv")

# Display the first few rows to verify the data
print(teams.head())

# Define the columns for different categories
offensive_stats = ['o_fgm', 'o_fga', 'o_ftm', 'o_fta', 'o_3pm', 'o_3pa', 'o_oreb', 'o_dreb', 'o_reb', 'o_asts', 'o_pf', 'o_stl', 'o_to', 'o_blk', 'o_pts']
defensive_stats = ['d_fgm', 'd_fga', 'd_ftm', 'd_fta', 'd_3pm', 'd_3pa', 'd_oreb', 'd_dreb', 'd_reb', 'd_asts', 'd_pf', 'd_stl', 'd_to', 'd_blk', 'd_pts']
team_rebounding = ['tmORB', 'tmDRB', 'tmTRB']
opponent_team_rebounding = ['opptmORB', 'opptmDRB', 'opptmTRB']
season_performance = ['won', 'lost', 'GP', 'homeW', 'homeL', 'awayW', 'awayL', 'confW', 'confL']

# Distribution of Offensive Statistics
for feature in offensive_stats:
    plt.figure(figsize=(10, 6))
    sns.histplot(teams[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

# Distribution of Defensive Statistics
for feature in defensive_stats:
    plt.figure(figsize=(10, 6))
    sns.histplot(teams[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

# Distribution of Team Rebounding
for feature in team_rebounding:
    plt.figure(figsize=(10, 6))
    sns.histplot(teams[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

# Distribution of Opponent Team Rebounding
for feature in opponent_team_rebounding:
    plt.figure(figsize=(10, 6))
    sns.histplot(teams[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

# Distribution of Season Performance
for feature in season_performance:
    plt.figure(figsize=(10, 6))
    sns.histplot(teams[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

# Box Plots to Identify Outliers in Offensive Statistics
for feature in offensive_stats:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=teams[feature])
    plt.title(f'Box Plot of {feature}')
    plt.ylabel(feature)
    plt.show()

# Box Plots to Identify Outliers in Defensive Statistics
for feature in defensive_stats:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=teams[feature])
    plt.title(f'Box Plot of {feature}')
    plt.ylabel(feature)
    plt.show()

# Correlation Heatmap for Offensive and Defensive Statistics
plt.figure(figsize=(16, 12))
correlation_matrix = teams[offensive_stats + defensive_stats].corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap for Offensive and Defensive Statistics')
plt.show()

# Team Performance Over Time
plt.figure(figsize=(14, 8))
sns.lineplot(data=teams, x='year', y='won', hue='tmID', marker='o')
plt.title('Team Wins Over Time')
plt.xlabel('Year')
plt.ylabel('Wins')
plt.legend(title='Team ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

plt.figure(figsize=(14, 8))
sns.lineplot(data=teams, x='year', y='lost', hue='tmID', marker='o')
plt.title('Team Losses Over Time')
plt.xlabel('Year')
plt.ylabel('Losses')
plt.legend(title='Team ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Playoff Status Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='playoff', data=teams)
plt.title('Playoff Status Distribution')
plt.xlabel('Playoff Status')
plt.ylabel('Count')
plt.show()

# Scatter Plots to Explore Relationships
plt.figure(figsize=(10, 6))
sns.scatterplot(x='o_pts', y='d_pts', data=teams, hue='playoff')
plt.title('Offensive Points vs Defensive Points')
plt.xlabel('Offensive Points')
plt.ylabel('Defensive Points')
plt.legend(title='Playoff Status')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='won', y='lost', data=teams, hue='playoff')
plt.title('Wins vs Losses')
plt.xlabel('Wins')
plt.ylabel('Losses')
plt.legend(title='Playoff Status')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='homeW', y='awayW', data=teams, hue='playoff')
plt.title('Home Wins vs Away Wins')
plt.xlabel('Home Wins')
plt.ylabel('Away Wins')
plt.legend(title='Playoff Status')
plt.show()

# Attendance Over Time
plt.figure(figsize=(14, 8))
sns.lineplot(data=teams, x='year', y='attend', hue='tmID', marker='o')
plt.title('Attendance Over Time')
plt.xlabel('Year')
plt.ylabel('Attendance')
plt.legend(title='Team ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
